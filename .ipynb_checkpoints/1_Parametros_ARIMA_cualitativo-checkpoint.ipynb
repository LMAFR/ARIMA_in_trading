```python
df_stocks_diff = train_data2-train_data2.shift()
df_stocks_diff.dropna(inplace=True)
adfuller(df_stocks_diff) # p-value = 0.0 (nulo dentro del límite de decimales, por lo visto) < 0.05. En consecuencia, 
                         # d=1 proporciona una serie estacionaria con un intervalo de confianza superior al 95%.
                         # Se ha obtenido el mismo resultado para ambos pares de divisas.
```
